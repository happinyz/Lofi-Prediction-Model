```python
artists_genre_series = pd.Series([x for _list in df["Artist Genre"] for x in _list])

genre_num = artists_genre_series.value_counts()[:15]
chart = sns.barplot(x = genre_num.index, y = genre_num)
chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
plt.show()
```
