```python
# Barchart of the song genres in the dataset


if type(df["Artist Genre"][0]) != list:
    df["Artist Genre"] = df["Artist Genre"].apply(eval)

artists_genre_series = pd.Series([x for _list in df["Artist Genre"] for x in _list])

genre_num = artists_genre_series.value_counts()[:15]

chart = sns.barplot(x = genre_num.index, y = genre_num)
chart.set_xticklabels(chart.get_xticklabels(), rotation = 90)
plt.show()
```

```python
key_features = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence']

# The descriptive stats of Lofi songs

df.loc[df["Is Lofi?"] == True, key_features].describe()
```

```python
# Let's generate the descriptive stats of non lofi songs
# Not particularly helpful since there are so many different genres in this dataset

df.loc[df["Is Lofi?"] == False, key_features].describe()
```

```python
# Is there any correlation between audio features in the lofi genre?

df.loc[df["Is Lofi?"] == True, key_features].corr()
```

```python
# Is there any correlation between audio features in non-lofi genres?
# Again, not particularly helpful since there are so many different genres in this dataset

df.loc[df["Is Lofi?"] == False, key_features].corr()
```

```python
# A heatmap and visualization of audio features correlation in lofi genres

cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(df.loc[df["Is Lofi?"] == True, key_features].corr(),
           vmax = .3, center = 0, cmap = cmap,
           square = True, linewidths = .5, cbar_kws = {"shrink": .5})
```
